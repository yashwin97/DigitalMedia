```python
#df_sochi_ad = pd.read_csv('./csv files/sochiAD.csv',names=["source","target","weight"],sep=" ")
df_sochi_ad = pd.read_csv('./sochi_A_full.csv',names=["source","target","weight"],sep="\t")
df_sochi_bl = pd.read_csv('./sochi_BL_full.csv',names=["source","target","weight"],sep="\t")
df_sochi_ex = pd.read_csv('./sochi_EX_full.csv',names=["source","target","weight"],sep="\t")


#df_sochi_ad.head()
#df_sochi_bl.head()
#df_sochi_ex.head()

#df_sochi_ad.dropna()
#df_sochi_bl.dropna()
#df_sochi_ex.dropna()
sochi_ad_sample = df_sochi_ad.sample(n=10000)
sochi_bl_sample = df_sochi_bl.sample(n=10000)
sochi_ex_sample = df_sochi_ex.sample(n=10000)
sochi_bl_sample.head()
```

```python
for index, row in sochi_ad_sample.iterrows():
    if not wordnet.synsets(row['source']):
        sochi_ad_sample.drop(index,inplace=True)
    elif not wordnet.synsets(row['target']):
        sochi_ad_sample.drop(index,inplace=True)
sochi_ad_sample['source'] = sochi_ad_sample['source'].str.lower()
sochi_ad_sample['target'] = sochi_ad_sample['target'].str.lower()
```

```python
for index, row in sochi_bl_sample.iterrows():
    if not wordnet.synsets(row['source']):
        sochi_bl_sample.drop(index,inplace=True)
    elif not wordnet.synsets(row['target']):
        sochi_bl_sample.drop(index,inplace=True)
sochi_bl_sample['source'] = sochi_bl_sample['source'].str.lower()
sochi_bl_sample['target'] = sochi_bl_sample['target'].str.lower()
```

```python
for index, row in sochi_ex_sample.iterrows():
    if not wordnet.synsets(row['source']):
        sochi_ex_sample.drop(index,inplace=True)
    elif not wordnet.synsets(row['target']):
        sochi_ex_sample.drop(index,inplace=True)
sochi_ex_sample['source'] = sochi_ex_sample['source'].str.lower()
sochi_ex_sample['target'] = sochi_ex_sample['target'].str.lower()
```

```python
G_s_ad = nx.from_pandas_edgelist(sochi_ad_sample,'source','target','weight')
G_s_bl = nx.from_pandas_edgelist(sochi_bl_sample,'source','target','weight')
G_s_ex = nx.from_pandas_edgelist(sochi_bl_sample,'source','target','weight')
```

```python
lp_ad = list(label_propagation_communities(G_s_ad))
lp_bl = list(label_propagation_communities(G_s_bl))
lp_ex = list(label_propagation_communities(G_s_ex))
```

```python
nx.write_gexf(G_s_ad, "test.gexf")
```

```python
degree_sequence_ad = sorted(G_s_ad.degree(), key=lambda x: x[1], reverse=True)
degree_sequence_bl = sorted(G_s_bl.degree(), key=lambda x: x[1], reverse=True)
degree_sequence_ex = sorted(G_s_ex.degree(), key=lambda x: x[1], reverse=True)
degree_sequence_ad[:10]
```

```python
Communities_AD = []
for fset in lp_ad:
    tempset = {}
    for key in fset:
        for degree in degree_sequence_ad:
            if key == degree[0]:
                tempset[key] = degree[1]
    t = {k: v for k, v in sorted(tempset.items(), key=lambda item: item[1], reverse=True)}
    Communities_AD.append(t)
Communities_AD
```

```python
Communities_BL = []
for fset in lp_bl:
    tempset = {}
    for key in fset:
        for degree in degree_sequence_bl:
            if key == degree[0]:
                tempset[key] = degree[1]
    t = {k: v for k, v in sorted(tempset.items(), key=lambda item: item[1], reverse=True)}
    Communities_BL.append(t)
Communities_BL
```
