```python
a=nx.pagerank(G_s_ad, alpha=0.9)
s = sorted(a.items(), key=lambda x: x[1], reverse=True)

for k, v in s:
    print(k, v)
```

```python
b=nx.pagerank(I, alpha=0.9)
d = sorted(b.items(), key=lambda x: x[1], reverse=True)

for k, v in d:
    print(k, v)
```

```python
#average of pagerank for real AD network
p=0
for d in a.values():
    p=p+d
print(p/len(a))    
```

```python
#average of pagerank for random AD network
p=0
for d in b.values():
    p=p+d
print(p/len(b))
```
